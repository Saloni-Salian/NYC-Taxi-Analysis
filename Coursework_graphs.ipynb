```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.patches as mpatches
```

```python
# Task 2 Part 1
data_for_task2_part1 = pd.read_csv('Task2/Part1.csv')
x_axis = data_for_task2_part1['Pickup_Borough']
y_axis = data_for_task2_part1['count']
plt.figure(figsize=(9,7))
plt.bar(x_axis, y_axis, color = 'yellow', edgecolor='black')
plt.title('Number of Yellow Taxi Pickups per Borough')
plt.xlabel('Yellow Taxi Pickup Boroughs')
plt.ylabel('Number of Trips')
plt.show()
```

```python
# Task 5
data_for_task5 = pd.read_csv('Task5/Task5_all.csv')
x_axis = data_for_task5['tpep_pickup_datetime']
y_axis = data_for_task5['count']
plt.figure(figsize=(13,7))
plt.scatter(x_axis, y_axis, color = 'yellow', edgecolor='black')
plt.title('Anomalies from Visualization')
plt.xlabel('Dates')
plt.ylabel('Number of Trips')
mean=93599.0
upperbound=(mean+mean*0.5)
lowerbound=(mean-mean*0.5)
plt.axhline(y=mean)
plt.axhline(y=upperbound, color='r')
plt.axhline(y=lowerbound, color='r')
plt.xticks(rotation=90)
plt.show()
```

```python
# Task 6
data_for_task6 = pd.read_csv('Task6/Task6.csv')
y_axis = data_for_task6['fare_per_mile']
plt.figure(figsize=(10,10))
plt.scatter(range(len(data_for_task6)), y_axis, color = 'yellow', edgecolor='black')
plt.title('Statistical Processing')
plt.ylabel('Fare Per Mile')
mean=10.478826871543824
maxPoint=data_for_task6['fare_per_mile'].max()
plt.axhline(y=mean)
plt.axhline(y=maxPoint, color='r')
plt.show()
```

```python
data_for_task9 = pd.read_csv('Task9/Task9.csv')
x_axis = data_for_task9['Borough']
y_axis = data_for_task9['count']
plt.figure(figsize=(9,7))
plt.bar(x_axis,y_axis, color= data_for_task9['colour'], edgecolor='black', log='True')
plt.xticks(rotation=90)
yellow_patch = mpatches.Patch(color='yellow', label='Number of Pickups for Yellow Taxis')
green_patch = mpatches.Patch(color='green', label='Number of Pickups for Green taxis')
plt.legend(handles=[yellow_patch,green_patch])
plt.title('Top 10 Most Active Boroughs') 
plt.xlabel('Borough')
plt.ylabel('Number of Trips')
plt.show() 
```
